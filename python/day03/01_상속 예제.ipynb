```python
# Phone 클래스를 추가
# 모델명 (model), 가격 (price)
# 기능: 생성자 - 모델명과 가격을 초기화, 모델조회(display)

# iPhone 클래스를 추가 (Phone 클래스를 상속)
# 페이스 아이디 적용 여부 (face_id)
# 기능 : 생성자 - 모델명, 가격, 페이스아이디 적용 여부를 초기화
# 모델조회 (display): 모델명, 가격, 페이스아이디 적용 여부부

# Galaxy 클래스를 추가 (Phone 클래스를 상속)
# 펜 지원 (s_pen)
# 기능: 생성자 - 모델명, 가격, 펜지원 여부를 초기화
# 모델조회 (display): 모델명, 가격, 펜지원 여부

# iPhone: iPhone15, 1250000, True
# iPhone: iPhoneSE, 650000, False
# Galaxy: S24 Ultra, 1690000, True
# Galaxy: Z Flip 5, 1390000, False

# 각 핸드폰을 조회

class Phone:
    def __init__(self, model, price):
        self.model = model
        self.price = price
    
    def display(self):
        return f'{self.model} - {self.price}'
    
class iPhone(Phone):
    def __init__(self, model, price, has_face_id):
        super(model, price)
        self.has_face_id = has_face_id
    
    def display(self):
        return f'{self.model} - {self.price} - {self.has_face_id}'


class iPhoneSE(Phone):
    def __init__(self, model, price, has_face_id):
        super(model, price)
        self.has_face_id = has_face_id
    def display(self):
        return f'{self.model} - {self.price} - {self.has_face_id}'

class GalaxyS24Ultra(Phone):
    def __init__(self, model, price, has_s_pen):
        super(model, price)
        self.has_s_pen = True
    def display(self):
        return f'{self.model} {self.price} {self.has_s_pen}'
    
class GalaxyZFlip5(Phone):
    def __init__(self, model, price, has_s_pen):
        super(model, price)
        self.has_s_pen = False
    def display(self):
        return f'{self.model} {self.price} {self.has_s_pen}'


myPhone = GalaxyS24Ultra("Galaxy S26 Utra", "1690000", True)
```
