```python
# 문자열.strip()
str1 = "       안녕하세요 \t\t\n\n"
print(str1)
print("==========")
print(str1.strip())
```

```python
# 문자열.split(구분자): 구분자를 기준으로 추출하여 리스트를 반환
str2 = "바나나 배 포도 사과"
fruits = str2.split()
print(fruits)

str3 = "바나나,배,포도,사과"
fruits2 = str3.split(',')
print(fruits2)
```

```python
# 문자열.join(리스트)
# 리스트 앞에 문자열들을 추가하여 하나의 문자열을 만듬
tags = ['AI', 'Python', 'Web']
result = ' #'.join(tags)
print(f"#{result}")
```

```python
# 문자열.startswith(A) : A로 시작하는지 알려줌
# 문자열.endswith(A) : A로 끝나는지 알려줌
filename = "report_수업자료.pdf"
print(f"{filename}은 보고서입니까? {filename.startswith("report")}")
print(f"{filename}은 pdf입니까? {filename.endswith("pdf")}")
```

```python
# 리스트 선언 []를 이용
fruits = ["바나나", "딸기", "포도", "배"]
print(fruits)
```

```python
# append() : 맨 뒤에 추가
fruits.append("딸기")
print(fruits)
```

```python
# remove() : 일치하는 첫번째 값을 삭제
fruits.remove("딸기")
print(fruits)
```

```python
# insert() : 번지에 값을 추가
fruits.insert(1, "오렌지")
print(fruits)
```

```python
# index() : 값이 몇번지에 있는지 알려줌
print(fruits.index("배"))
```

```python
# len() : 리스트 길이
print(len(fruits))
```

```python
# 튜플 선언 : ()
fruits = ("사과", "배", "바나나")
print(fruits)

# 튜플에서 append
# fruits.append("딸기") # 에러 발생. 튜플은 수정 불가

# 튜플에서 len
print(len(fruits))

print(fruits)
# list() : 튜플을 리스트로 변환 
fruits2 = list(fruits)
# fruits2는 리스트이기 때문에 수정이 가능
fruits2.append("딸기")
print(fruits2)
# tuple() : 튜플을 반환
fruits = tuple(fruits2)
print(fruits)
```

```python
# 딕셔너리 선언 {} 
dic = {
    "name" : "홍길동",
    "age" : 21,
    "addr" : "서울시"
}
print(dic)
# 이름 확인
print(f"이름 : {dic["name"]}")
# print(f"이름 : {dic["name1"]}") # 키가 없는 경우 에러 발생
print(f"이름 : {dic.get("name", "익명")}")
print(f"이름 : {dic.get("name1", "익명")}") # 키가 없어도 기본값이 실행
# 새로운 키 추가
dic["job"] = "강사"
print(dic)
```

```python
# set : {}를 선언
set_a = { "바나나", "사과", "딸기", "포도", "바나나" }
print(set_a)
set_b = { "바나나", "귤", "오렌지", "딸기" }
print(set_b)
```

```python
# 합집합 : |, union()
# "바나나", "딸기", "포도", "사과", "귤", "바나나", "오렌지", "딸기"
# => "바나나", "딸기", "포도", "사과", "귤", "오렌지"
print("A와 B의 합집합")
print(set_a | set_b)
print(set_a.union(set_b))
```

```python
# 교집합 : &, intersection()
print("A와 B의 교집합 : 두 집합에 공통으로 있는 값들")
print(set_a & set_b)
print(set_a.intersection(set_b))
```

```python
# 차집합: -, difference()
print("A - B : 집합 B에 없고 A에만 있는 값들")
print(set_a - set_b)
print(set_a.difference(set_b))
```

```python
# 대칭 차집합: ^, symmetric_difference()
# 겹치지 않은 모든 원소들 모음
# (A - B) | (B - A)
# (A | B) - (A & B)
print(set_a ^ set_b)
print(set_a.symmetric_difference(set_b))
```

```python
# 집합 예제 (조건, 이름은 중복 X)
# 수학을 듣는 학생 목록
course_math = {"홍길동", "임꺽정", "둘리", "고길동", "하니"}
# 국어를 듣는 학생 목록
course_kor = {"이순신", "이도", "김삿갓", "둘리", "홍길동"}

#수학만 듣는 학생 조회
print(f"수학만 듣는 학생 목록: {course_math - course_kor}")

#수학 국어 모두 듣는 학생 목록 조회
print(f"수학 국어 모두 듣는 학생 목록: {course_math & course_kor}")

#수학 또는 국어를 듣는 모든 학생의 목록
print(f"수학 또는 국어를 듣는 모든 학생 목록: {course_math | course_kor}")

#수학이나 국어를 하나만 듣는 학생 목록을 조회
print(f"수학이나 국어를 하나만 듣는 학생 목록: {course_math ^ course_kor}")
```
