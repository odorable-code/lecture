```python
# 붗꽃 예제를 딥러닝으로 예측
from tensorflow.keras import models, layers
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
# 붓꽃 데이터 생
from sklearn.datasets import load_iris

data = load_iris() 
```

```python
# 독립, 종속 분리

X = data['data']
y = data['target']
```

```python
# 훈련, 테스트 분리

from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 정규화
normalizer = layers.Normalization(axis=-1)
normalizer.adapt(X_train) 
```

```python
# 모델 설계
model = models.Sequential([
    normalizer,
    layers.Dense(16, activation='relu'),
    layers.Dense(8, activation='relu'),
    layers.Dense(3, activation='softmax') #출력
])
```

```python
# 모덷 설정
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 학습
history = model.fit(X_train, y_train, epochs=1000, batch_size=16, verbose=0)
```

```python
# 실제 (100번지) 6.3, 3.3, 6, 2.5 => 2
# 예측 6.4, 3.2, 6, 2.4 => 2 
new_iris = np.array([[6.4, 3.2, 6, 2.4]], dtype='float32')
prediction = model.predict(new_iris)
```
