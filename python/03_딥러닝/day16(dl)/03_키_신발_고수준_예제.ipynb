```python
# нӮӨ лҚ°мқҙн„°мҷҖ мӢ л°ң лҚ°мқҙн„°лҘј мқҙмҡ©н•ҳм—¬ н•ҷмҠөн•ҳкі , нӮӨ лҚ°мқҙн„°лҘј мқҙмҡ©н•ҳм—¬ мӢ л°ң мӮ¬мқҙмҰҲлҘј мҳҲмёЎн•ҳлҠ” мҳҲм ң
# кі мҲҳмӨҖмңјлЎң мһ‘м„ұ => Keras лӘЁлҚё мқҙмҡ©
import tensorflow as tf
нӮӨлҚ°мқҙн„° = tf.constant([170, 180, 175, 160], dtype=tf.float32)
мӢ л°ңлҚ°мқҙн„° = tf.constant([260, 270, 265, 255], dtype=tf.float32)
```

```python
# лӘЁлҚё м„Өкі„
model = tf.keras.Sequential([
    # л Ҳмқҙм–ҙлҘј мһ‘м„ұ
    # л§Ҳм§Җл§ү л Ҳмқҙм–ҙлҠ” м¶ңл Ҙмёө
    # 1мқҖ н•ҙлӢ№ л Ҳмқҙм–ҙлҘј м§ҖлӮҳл©ҙ кІ°кіјк°Җ лӘҮк°ң лӮҳмҷҖм•ј н•ҳлҠ”м§ҖлҘј кІ°м •\
    # м§ҖкёҲмқҖ мӢ л°ң мӮ¬мқҙмҰҲ к°’ 1к°ңл§Ң лӮҳмҳӨл©ҙ лҗҳлӢҲк№Ң 1
    # input_shape = (1,): мһ…л Ҙмёөмқҳ нҳ•нғң, мһ…л ҘмёөмқҖ нӮӨ 1к°ңмқҙлӢҲк№Ң 1
    tf.keras.layers.Dense(1, input_shape=(1,))
])

# лӘЁлҚё м„Өм •, мҳөнӢ°л§ҲмқҙлҘј м–ҙл–Ө лӘЁлҚёлЎң н•  кұҙм§Җ, мҳӨм°ЁлҠ” м–ҙл–Ө л°©мӢқмңјлЎң кі„мӮ°н•  кІғмқём§Җ кІ°м •
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1), loss='mse')
```

```python
# н•ҷмҠө мӢӨн–ү
model.fit(нӮӨлҚ°мқҙн„°, мӢ л°ңлҚ°мқҙн„°, epochs=1000, verbose=0)
```

```python
import numpy as np
мёЎм •нӮӨ = 173
# н…җм„ң лӘЁлҚёл“ӨмқҖ мҳҲмёЎн•  л•Ң, numpy лҰ¬мҠӨнҠёк°Җ н•„мҡ”
мҳҲмёЎмӢ л°ң = model.predict(np.array([[мёЎм •нӮӨ]], dtype='float32'))
print(f'мёЎм •нӮӨ : {мёЎм •нӮӨ}, мҳҲмёЎмӢ л°ңмӮ¬мқҙмҰҲ : {мҳҲмёЎмӢ л°ң[0][0]:.2f}')
```
