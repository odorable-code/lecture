```python
# 샘플 데이터를 학습해서 30도일 때 아이스크림 판매량을 예측하세요
# 그래프도 출력

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('선형회귀_아이스크림.csv')
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred30 = model.predict([[30]] )
print(f'30도일때 판매량 {y_pred30.round(0)}') 
```

```python
score = model.score(X_test, y_test)
print(f'성능: {score}')
```

```python
# 데이터 시각화로 훈련 데이터, 테스트데이터, 예측 직선을 확인
# 학습 결과를 그래프로 출력
plt.scatter(X_train, y_train, color='blue') # 훈련 데이터를 점으로 출력
plt.plot(X_train, model.predict(X_train), color='red') # 학습을 통해 나온 직선
plt.scatter(X_test, model.predict(X_test), color='gray') # 테스트 데이터의 예측 데이터
plt.scatter(X_test, y_test, color='green') # 테스트 데이터의 실제 데이터
plt.title('sales by temp') # 그래프 제목
plt.xlabel('temp')# x축 => 독립변수
plt.ylabel('sales')# y축 => 종속변수
plt.show() # 그래프 출력
```
